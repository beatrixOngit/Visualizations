##### Matplotlib Colors

There are many possible ways to identify a color in matplotlib. Read the color documentation to see all 8 ways to specify a color. Here are the full color names with the shade they represent

```python
fig, ax = plt.subplots()
x = np.arange(-5, 5)
y = x ** 2
ax.plot(x, y, color='darkblue', linestyle='--', marker='s')
ax.set_facecolor('azure')
```

**set_facecolor**: is the new version of set_axis_bgdcolor and sets the background color of the plot

##### About LinesStyles
The default setting is a **solid line**. However, in case of multiple lines on the same plot, changing the linestyle might be better for readability. Use pandas to scrape the kinds of linestyle that exist within **Matplotib**. 

```python
dfs = pd.read_html('https://matplotlib.org/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D')
dfs[0]
```

**Note:** The following snippet covers a lot of the points we've discussed above:


```python
fig, ax = plt.subplots()

ax.plot(x, y, color='purple', linestyle = "-.", marker='s', mec='black', mew = 1, mfc ='gray')
ax.set_facecolor("aliceblue")
```

##### Markers
Notice the argument 'marker' in the above code. A marker indicates a single data point on the plot. The argument marker can let you choose how each data point looks like. This is specially helpful when you are trying to draw attention to where each data point falls more than the trend of the line. The later arguments, mec='black', mew = 1, mfc ='gray'. These stand for 
* marker edge color
* marker edge width, and
* marker face color

So, tons of ways to customize the data points themselves.
